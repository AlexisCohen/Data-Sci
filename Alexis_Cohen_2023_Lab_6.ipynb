```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
import urllib.request 
```

```python
url = "https://raw.githubusercontent.com/AlexisCohen/Data-Sci/main/cereal_data.csv"
#imports the url from github into the variable url
```

```python
urllib.request.urlretrieve(url,'cereal_data.csv')
#this extracts the data from the inserted url
```

```python
cereal_data = pd.read_csv('cereal_data.csv')  
#this puts the data extracted into the variable cereal_data
```

```python
x = cereal_data[["sugars"]]
y = cereal_data[["calories"]]

plt.plot(x,y,'.')
plt.xlabel("sugars")
plt.ylabel("calories")
plt.title("sugars vs calories")

#This is the same graph with labels on the x and y axis as well as a title
```

```python
x = cereal_data[["fat"]]
y = cereal_data[["calories"]]

plt.plot(x,y,'.')
plt.xlabel("fat")
plt.ylabel("calories")
plt.title("fats vs calories")

#This creates a graph of fat and calories with labels and a title
```

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression() 
```

```python
X = cereal_data[["fat"]]
Y = cereal_data[["calories"]]
#assigns values to the X and Y variables
```

```python
linear_regressor.fit(X, Y)
Y_pred = linear_regressor.predict(X)
#this assigns a variable to predict what the best line through the graph
```

```python
plt.plot(X, Y,'.')
plt.plot(X, Y_pred, color='blue')
plt.show()

#this is the fat and calories graph with a line through the 
#middle that is as close to eqidistant from all the points
```

```python
m_guess = 1
b_guess = 1

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');

#the red line is the guess and the green is the original with noise
```

```python
m_guess = 5
b_guess = 15

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
#this is the graph with noise with the guess being higher
```

```python
loss = np.sum((y_guess - y)**2)
#this variable calculates the loss/difference from the original function
```

```python
for m_guess in np.linspace(-10,10,10):
    for b_guess in np.linspace(-10,10,10):
        
        y_guess = m_guess * x + b_guess

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y,'g.');
        plt.show()

        loss = np.sum((y_guess - y)**2)

        print(loss,m_guess,b_guess)

#this code goes through all possible guesses for the middle line
```

```python
best_loss = 100000
best_m = 0
best_b = 0 

for m_guess in np.linspace(-10,10,20):
    for b_guess in np.linspace(-10,10,20):
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 

            plt.plot(x,y_guess,'r.');
            plt.plot(x,y,'g.');
            plt.show()

            print(loss,m_guess,b_guess)

#this gives the previous code more attempts to find the most accurate guess and then prints them
```
