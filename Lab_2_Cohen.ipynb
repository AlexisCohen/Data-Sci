#Pie Charts

```python
import matplotlib.pyplot as plt
```

```python
labels = ['Frogs','Dogs','Cats',"Birds"]

sizes = [15,3,45,10]

explode = (0,0.1,0,0) # depending on were the 0.1 is corolates to the 4 values in both of the lists and which section gets extruded

fig1,ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')

#this is how you make a pie chart using the matplotlib import
```

```python
labels = ['cakes','cookies','crepes',"grapes"]

sizes = [67,25,2,89]

explode = (0,0,0.1,0)

fig1,ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
```

#Functions

```python
def flip_coin():
  coin = np.random.random()>0.5
  if coin == True:
    print("Heads")
  else:
    print("Tails")

# now it is a function, the code no longer has to be repeated to execute it in a seperate instance
```

```python
flip_coin()
```

```python
for i in range(10):
  flip_coin()
```
